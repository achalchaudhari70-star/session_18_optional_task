```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/insurance (1).csv")
```

```python
print("First 10 Rows:")
df.head(10)
```

```python
print("\nDataset Shape:")
df.shape
```

```python
print("\nData Types:")
df.dtypes
```

```python
print("Missing Values in Each Column:")
print(df.isnull().sum())
```

```python
print("\nNumber of Duplicate Rows:")
print(df.duplicated().sum())
print("\nRemove duplicate rows:")
df = df.drop_duplicates()
print("\nShape After Removing Duplicates:")
print(df.shape)
```

```python
print("Statistical Summary:")
df.describe()
```

```python
def plot_histogram(column):
    plt.figure(figsize=(6,4))
    plt.hist(df[column], bins=20,edgecolor="black")
    plt.title("Histogram of " + column)
    plt.xlabel(column)
    plt.ylabel("Frequency")
    plt.show()
    plt.tight_layout()

plot_histogram("age")
plot_histogram("bmi")
plot_histogram("expenses")
```

```python
# Overall Observation (Histograms)
# The Age distribution is fairly uniform, with most policyholders between 20 and 40 years.
# The BMI values are concentrated between 25 and 35, forming an approximately bell-shaped distribution.
# The Children column shows that most policyholders have 0 or 1 child, while fewer have larger families.
# The Expenses distribution is right-skewed, indicating that most people have lower medical expenses, while a small number have very high insurance charges.
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df['age'], df['expenses'])
plt.title("Scatter Plot of Age vs Expenses")
plt.xlabel("Age")
plt.ylabel("Insurance Expenses")
```

```python
# Observation
# The scatter plot shows a positive relationship between age and insurance expenses.
# As age increases, insurance expenses generally increase.
# Some people have very high insurance expenses, appearing as outliers.
# The plot indicates that age influences insurance charges, but other factors such as BMI and smoking status also affect the expenses.
```

```python
plt.figure(figsize=(6,4))
plt.boxplot(df['expenses'],patch_artist=True, boxprops=dict(facecolor='lightgreen'))
plt.title("Box Plot of Insurance Expenses")
plt.ylabel("Expenses")
```

```python
# Observation
# The box plot shows that insurance expenses have a wide range of values.
# Several outliers are present above the upper whisker, indicating a few policyholders have very high medical expenses.
# Most expenses are concentrated within a lower range.
# The distribution is positively (right) skewed due to the high-value outliers.
```

```python
def count_plot(column, title):
    plt.figure(figsize=(6,4))
    sns.countplot(x=column, data=df, palette="Set2")
    plt.title(title)
    plt.xlabel(column)
    plt.ylabel("Count")
    plt.show()

# Count Plots
count_plot("sex", "Count Plot of Sex")
count_plot("smoker", "Count Plot of Smoker")
count_plot("region", "Count Plot of Region")
```

```python
# Observation
# The number of male and female policyholders is almost equal.
# Most policyholders are non-smokers, while smokers are fewer.
# The four regions have a fairly balanced number of policyholders, with no region dominating the dataset.
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df[['age', 'bmi', 'children', 'expenses']].corr(),annot=True,cmap="coolwarm")
plt.title("Correlation Heatmap")
```

```python
# Observation
# Age has a moderate positive correlation with expenses.
# BMI has a weak positive correlation with expenses.
# Children has a very weak correlation with the other variables.
# The heatmap indicates that age is the most influential numeric feature for predicting insurance expenses,
# while children has the least impact.
```

```python
plt.figure(figsize=(6,6))
df['smoker'].value_counts().plot(kind='pie',autopct='%1.1f%%',colors=['#FF6B6B', '#4ECDC4'])
plt.title("Distribution of Smokers")
plt.ylabel("")
```

```python
# Observation
# The majority of policyholders are non-smokers.
# A smaller percentage of policyholders are smokers.
# The pie chart clearly shows that non-smokers make up the largest share of the dataset.
```

```python
print("Before One-Hot Encoding:")
print(df.head())
df_encoded = pd.get_dummies(df,columns=['sex', 'smoker', 'region'],drop_first=True)
print("\nAfter One-Hot Encoding:")
df_encoded.head()
```

```python
# Observation
# Before Encoding: The dataset contains categorical columns (sex, smoker, and region) with text values.
# After Encoding: These categorical columns are converted into multiple numeric columns containing 0 and 1 values.
# The encoded dataset can now be used for machine learning algorithms that require numerical input.
```

```python
from sklearn.preprocessing import StandardScaler
# Separate input and output features
X = df_encoded.drop('expenses', axis=1)
y = df_encoded['expenses']

# Before Scaling
print("Before Standard Scaling:")
print(X.head())

# Standard Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Convert to DataFrame
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

# After Scaling
print("\nAfter Standard Scaling:")
print(X_scaled.head())
```

```python
# Observation
# Before scaling, the input features have different ranges of values.
# After scaling, all input features have a mean close to 0 and a standard deviation close to 1.
# Standard Scaling ensures that all features contribute equally during model training and
# improves the performance of many machine learning algorithms
```
