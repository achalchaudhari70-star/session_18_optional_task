```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/netflix_dataset.csv")
```

```python
print("First 10 Rows:")
df.head(10)
```

```python
print("\nDataset Shape:")
df.shape
```

```python
print("\nData Types:")
df.dtypes
```

```python
print("Missing Values in Each Column:")
df.isnull().sum()
```

```python
print("\nNumber of Duplicate Rows:")
print(df.duplicated().sum())
print("\nRemove duplicate rows:")
df = df.drop_duplicates()
print("\nShape After Removing Duplicates:")
df.shape
```

```python
print("Statistical Summary:")
df.describe()
```

```python

def histogram(column,color):
    plt.figure(figsize=(6,4))
    sns.histplot(df[column], bins=10, color=color, edgecolor="black")
    plt.title(f"Histogram of {column}")
    plt.xlabel(column)
    plt.ylabel("Frequency")
    plt.show()

# Function Calls
histogram("release_year","skyblue")
histogram("rating","red")
histogram("duration_min","orange")
histogram("views_millions","pink")
```

```python
# Observations
# Release Year: Most titles are from recent years.
# Rating: Most titles have good ratings.
# Duration (Minutes): Most titles have a medium duration.
# Views (Millions): Most titles have average views, while only a few have very high views.
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(y=df["rating"], color="tomato")
plt.title("Box Plot of Rating")
plt.ylabel("Rating")
plt.show()
```

```python
# Observation
# Most ratings are within a similar range.
# There are few or no outliers in the ratings
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df['rating'], df['views_millions'])
plt.title("Scatter Plot of rating vs views_millions")
plt.xlabel("rating	")
plt.ylabel("views_millions")
```

```python
# Observation:

# There is no strong relationship between rating and views_millions.
# High-rated content may have higher views, but it is not always the case.
# Views depend on factors other than rating, such as popularity and audience interest
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df[[ 'release_year', 'rating', 'duration_min','views_millions']].corr(),annot=True,cmap="viridis")
plt.title("Correlation Heatmap")
```

```python
# Observation:

# The heatmap shows the correlation between numerical features.
# Values close to +1 indicate a positive relationship.
# Values close to -1 indicate a negative relationship.
# Most features show weak correlation with each other.
# Correlation helps identify relationships between variables.
```

```python
def count_plot(column):
    plt.figure(figsize=(6,4))
    sns.countplot(x=column, data=df, palette="Set2")
    plt.title(f"Histogram of {column}")
    plt.xlabel(column)
    plt.ylabel("Count")
    plt.show()

# Count Plots
count_plot("type")
count_plot("genre")
count_plot("country")
```

```python
# Observation:

# Type:

# Shows the count of different content types.
# One type has more entries than the other.

# Genre:

# Shows the number of titles in each genre.
# Some genres are more common than others.

# Country:

# Shows content count by country.
# A few countries have higher content availability.
```

```python
plt.figure(figsize=(6,6))
df['country'].value_counts().plot(kind='pie',autopct='%1.1f%%',colors=['#FF6B6B', '#4ECDC4'],startangle=90)
plt.title("Distribution of country")
plt.ylabel("")
```

```python
print("Before One-Hot Encoding:")
print(df.head())
df_encoded = pd.get_dummies(df,columns=['show_id', 'title', 'type','genre','country'],drop_first=True)
print("\nAfter One-Hot Encoding:")
print(df_encoded.head())
```

```python
# Observation:

# Before encoding, categorical data is in text format.
# After encoding, text values are converted into numerical (0 and 1) format.
# The dataset is ready for machine learning.
```

```python
from sklearn.preprocessing import StandardScaler
# Separate input and output features
X = df_encoded.drop('views_millions', axis=1)
y = df_encoded['views_millions']

# Before Scaling
print("Before Standard Scaling:")
print(X.head())

# Standard Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Convert to DataFrame
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

# After Scaling
print("\nAfter Standard Scaling:")
print(X_scaled.head())
```

```python
# Observation:

# Before scaling, features have different ranges of values.
# After Standard Scaling, values are converted to a common scale.
# Scaled data improves performance of machine learning models.
```
